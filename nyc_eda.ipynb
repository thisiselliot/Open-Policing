```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('data/2014.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df = df[df.xcoord!=' ']
df = df[df.ycoord!=' ']
```

```python
df['xcoord'] = df.xcoord.astype(int)
df['ycoord'] = df.ycoord.astype(int)
```

```python
df.ycoord.dtype
```

```python
plt.figure(figsize=(20,12))
sns.scatterplot(x=df.xcoord, y=df.ycoord, alpha=0.1, hue=df.race)
```
